```python
# ==============================================================================
# BAGIAN 0: INSTALASI DAN IMPORT LIBRARY (MENGGUNAKAN PYTORCH UNTUK BERT)
# ==============================================================================
# Hapus instalasi lama
!pip uninstall -y transformers tensorflow tokenizers torch

# Instal PyTorch (versi kompatibel dengan CUDA 12.1 di GPU T4)
!pip install torch torchvision torchaudio --index-url https://download.pytorch.org/whl/cu121

# Instal library lainnya
!pip install transformers tensorflow scikit-learn

# Import semua library yang akan digunakan
import re
import string
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm.notebook import tqdm

# Library untuk Dataset
from datasets import load_dataset, DatasetDict

# Library untuk BERT (Sekarang menggunakan PyTorch)
import torch
from transformers import BertTokenizer, BertForSequenceClassification

# Library untuk LSTM dan model ML (Masih menggunakan TensorFlow)
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout, SpatialDropout1D
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

# Library untuk Metrik
from sklearn.metrics import (
    confusion_matrix,
    roc_auc_score,
    precision_score,
    recall_score,
    f1_score,
    classification_report
)

print("✅ Semua library berhasil diimpor.")
print(f"Versi PyTorch: {torch.__version__}")
print(f"Versi TensorFlow: {tf.__version__}")
print(f"GPU PyTorch Tersedia: {torch.cuda.is_available()}")
print(f"GPU TensorFlow Tersedia: {len(tf.config.list_physical_devices('GPU')) > 0}")

# ==============================================================================
# BAGIAN 1: LOAD DAN PREPROCESSING DATA AWAL
# ==============================================================================
# Load dataset
dataset = load_dataset("drorrabin/phishing_emails-data")

# Fungsi untuk mengubah label string menjadi integer
def label_to_int(example):
    if example['email_type'] == 'phishing email':
        example['label'] = 1  # Phishing
    elif example['email_type'] == 'safe email':
        example['label'] = 0  # Legitimate
    else:
        example['label'] = -1 # Beri label error
    return example

# Terapkan konversi ke seluruh dataset
dataset = dataset.map(label_to_int)

# Fungsi untuk membersihkan teks email
def clean_text(text):
    opening_pattern = re.compile(r"is the following email safe or phishing\?\?.*\n", re.IGNORECASE)
    text = opening_pattern.sub("", text)
    closing_pattern = re.compile(r"\s*(?:Email type is|email_type is):.*$", re.IGNORECASE)
    text = closing_pattern.sub("", text)
    header_pattern = re.compile(r"(date|sender|receiver|email subject|email body):\s*", re.IGNORECASE)
    text = header_pattern.sub("", text)
    text = text.lower()
    text = re.sub(r'\[.*?\]', '', text)
    text = re.sub(r'<.*?>+', '', text)
    text = re.sub(r'[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub(r'\n', ' ', text)
    text = re.sub(r'\s+', ' ', text).strip()
    text = re.sub(r'\w*\d\w*', '', text)
    text = re.sub(r'http\S+|www\S+', '', text)
    return text

# Terapkan pembersihan
dataset = dataset.map(lambda x: {'text_clean': clean_text(x['text'])})

# ==============================================================================
# BAGIAN 2: PEMBAGIAN DATA (TRAIN & TEST SAJA)
# ==============================================================================
source_dataset = dataset['train']
full_dataset = source_dataset.train_test_split(test_size=0.2, seed=42)
final_dataset = DatasetDict({
    'train': full_dataset['train'],
    'test': full_dataset['test']
})

print("\nStruktur Dataset Setelah Pembagian Ulang (80% Train, 20% Test):")
print(final_dataset)
print(f"\nUkuran Train Set: {len(final_dataset['train'])}")
print(f"Ukuran Test Set: {len(final_dataset['test'])}")

X_train_raw = final_dataset['train']['text_clean']
y_train = np.array(final_dataset['train']['label'])
X_test_raw = final_dataset['test']['text_clean']
y_test = np.array(final_dataset['test']['label'])

# ==============================================================================
# BAGIAN 3: MODEL 1 - BERT (MENGGUNAKAN PYTORCH)
# ==============================================================================
print("\n" + "="*80)
print("🤖 MEMULAI PELATIHAN MODEL 1: BERT (dengan PyTorch)")
print("="*80)

# --- 3.1 Tokenisasi untuk BERT ---
bert_tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

def tokenize_bert(examples):
    return bert_tokenizer(
        examples['text_clean'],
        padding='max_length',
        truncation=True,
        max_length=256
    )

tokenized_bert_train = final_dataset['train'].map(tokenize_bert, batched=True)
tokenized_bert_test = final_dataset['test'].map(tokenize_bert, batched=True)

# --- 3.2 Siapkan format untuk PyTorch ---
tokenized_bert_train.set_format('torch', columns=['input_ids', 'attention_mask', 'label'])
tokenized_bert_test.set_format('torch', columns=['input_ids', 'attention_mask', 'label'])

# Buat DataLoader untuk memudahkan batching
from torch.utils.data import DataLoader

bert_train_dataloader = DataLoader(tokenized_bert_train, shuffle=True, batch_size=16)
bert_test_dataloader = DataLoader(tokenized_bert_test, batch_size=16)

# --- 3.3 Build Model BERT dengan PyTorch ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
bert_model_pytorch = BertForSequenceClassification.from_pretrained(
    'bert-base-uncased',
    num_labels=2
)
bert_model_pytorch.to(device)

# --- 3.4 Training BERT dengan PyTorch ---
optimizer = torch.optim.AdamW(bert_model_pytorch.parameters(), lr=3e-5)
epochs = 3

print("🚀 Memulai proses training BERT dengan PyTorch...")
for epoch in range(epochs):
    bert_model_pytorch.train()
    total_loss = 0
    for batch in tqdm(bert_train_dataloader, desc=f"Epoch {epoch+1}/{epochs}"):
        optimizer.zero_grad()

        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels = batch['label'].to(device)

        outputs = bert_model_pytorch(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
        loss = outputs.loss
        total_loss += loss.item()

        loss.backward()
        optimizer.step()

    avg_train_loss = total_loss / len(bert_train_dataloader)
    print(f"Epoch {epoch+1} | Average Training Loss: {avg_train_loss:.4f}")
print("✅ Proses training BERT selesai!")

# --- 3.5 Prediksi dengan BERT PyTorch ---
def get_bert_pytorch_predictions(dataloader, model):
    model.eval()
    y_true = []
    y_pred_proba = []
    with torch.no_grad():
        for batch in tqdm(dataloader, desc="Memprediksi dengan BERT PyTorch"):
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            labels = batch['label'].to(device)

            outputs = model(input_ids=input_ids, attention_mask=attention_mask)
            logits = outputs.logits
            probabilities = torch.nn.functional.softmax(logits, dim=1)[:, 1].cpu().numpy()

            y_true.extend(labels.cpu().numpy())
            y_pred_proba.extend(probabilities)

    y_pred_class = [1 if p > 0.5 else 0 for p in y_pred_proba]
    return np.array(y_true), np.array(y_pred_class), np.array(y_pred_proba)

y_test_bert_true, y_test_bert_pred, y_test_bert_proba = get_bert_pytorch_predictions(bert_test_dataloader, bert_model_pytorch)

# ==============================================================================
# BAGIAN 4: MODEL 2 - LSTM (TIDAK BERUBAH, MASIH PAKE TF)
# ==============================================================================
print("\n" + "="*80)
print("🧠 MEMULAI PELATIHAN MODEL 2: LSTM")
print("="*80)

# --- 4.1 Tokenisasi dan Padding untuk LSTM ---
lstm_tokenizer = Tokenizer(num_words=5000, oov_token='<OOV>')
lstm_tokenizer.fit_on_texts(X_train_raw)

X_train_seq = lstm_tokenizer.texts_to_sequences(X_train_raw)
X_test_seq = lstm_tokenizer.texts_to_sequences(X_test_raw)

maxlen = 200
X_train_pad = pad_sequences(X_train_seq, maxlen=maxlen)
X_test_pad = pad_sequences(X_test_seq, maxlen=maxlen)

# --- 4.2 Build dan Compile Model LSTM ---
lstm_model = Sequential()
lstm_model.add(Embedding(input_dim=5000, output_dim=128, input_length=maxlen))
lstm_model.add(SpatialDropout1D(0.3))
lstm_model.add(LSTM(100, dropout=0.3, recurrent_dropout=0.3))
lstm_model.add(Dense(64, activation='relu'))
lstm_model.add(Dropout(0.5))
lstm_model.add(Dense(1, activation='sigmoid'))

lstm_model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# --- 4.3 Training LSTM ---
print("🚀 Memulai proses training LSTM...")
lstm_history = lstm_model.fit(
    X_train_pad, y_train,
    epochs=5,
    batch_size=64,
    verbose=1
)
print("✅ Proses training LSTM selesai!")

# --- 4.4 Prediksi dengan LSTM ---
y_test_lstm_pred_proba = lstm_model.predict(X_test_pad).ravel()
y_test_lstm_pred = (y_test_lstm_pred_proba > 0.5).astype("int32")

# ==============================================================================
# BAGIAN 5: MODEL 3 - TF-IDF + LOGISTIC REGRESSION (TIDAK BERUBAH)
# ==============================================================================
print("\n" + "="*80)
print("📊 MEMULAI PELATIHAN MODEL 3: TF-IDF + Logistic Regression")
print("="*80)

# --- 5.1 Ekstraksi Fitur TF-IDF ---
tfidf_vectorizer = TfidfVectorizer(max_features=5000, ngram_range=(1, 2))
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train_raw)
X_test_tfidf = tfidf_vectorizer.transform(X_test_raw)

# --- 5.2 Training Model Logistic Regression ---
print("🚀 Memulai proses training Logistic Regression...")
lr_model = LogisticRegression(solver='liblinear', random_state=42)
lr_model.fit(X_train_tfidf, y_train)
print("✅ Proses training Logistic Regression selesai!")

# --- 5.3 Prediksi dengan TF-IDF + LR ---
y_test_lr_pred = lr_model.predict(X_test_tfidf)
y_test_lr_pred_proba = lr_model.predict_proba(X_test_tfidf)[:, 1]

# ==============================================================================
# BAGIAN 6: EVALUASI DAN PERBANDINGAN SEMUA MODEL
# ==============================================================================
print("\n" + "="*80)
print("📈 EVALUASI DAN PERBANDINGAN KINERJA MODEL")
print("="*80)

def calculate_metrics(y_true, y_pred, y_proba):
    return {
        'Accuracy': np.mean(y_true == y_pred),
        'Precision': precision_score(y_true, y_pred),
        'Recall': recall_score(y_true, y_pred),
        'F1-Score': f1_score(y_true, y_pred),
        'AUC': roc_auc_score(y_true, y_proba)
    }

metrics_bert = calculate_metrics(y_test_bert_true, y_test_bert_pred, y_test_bert_proba)
metrics_lstm = calculate_metrics(y_test, y_test_lstm_pred, y_test_lstm_pred_proba)
metrics_lr = calculate_metrics(y_test, y_test_lr_pred, y_test_lr_pred_proba)

df_comparison = pd.DataFrame({
    'BERT (PyTorch)': metrics_bert,
    'LSTM (TensorFlow)': metrics_lstm,
    'TF-IDF + LR': metrics_lr
}).T

df_comparison_formatted = df_comparison.applymap('{:.4f}'.format)
print("Tabel Perbandingan Metrik pada Test Set:")
print(df_comparison_formatted.to_string())

print("\n" + "="*80)
print("📊 Visualisasi Perbandingan Metrik")
print("="*80)

df_plot = df_comparison.reset_index().melt(id_vars='index', var_name='Metrik', value_name='Nilai')
df_plot.rename(columns={'index': 'Model'}, inplace=True)

plt.figure(figsize=(14, 8))
ax = sns.barplot(x='Metrik', y='Nilai', hue='Model', data=df_plot, palette='magma')
ax.set_title('Perbandingan Metrik Kinerja: BERT vs. LSTM vs. TF-IDF + LR', fontsize=16)
ax.set_ylabel('Nilai Skor', fontsize=12)
ax.set_xlabel('Metrik', fontsize=12)
ax.set_ylim(0, 1.05)

for p in ax.patches:
    ax.annotate(f"{p.get_height():.3f}",
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center',
                xytext=(0, 9),
                textcoords='offset points')

plt.legend(title='Model', fontsize=12)
plt.tight_layout()
plt.show()

target_names = ['Legitimate (Aman)', 'Phishing (Berbahaya)']
print("\n" + "="*40 + " BERT (PyTorch) " + "="*40)
print(classification_report(y_test_bert_true, y_test_bert_pred, target_names=target_names))
print("\n" + "="*40 + " LSTM (TensorFlow) " + "="*40)
print(classification_report(y_test, y_test_lstm_pred, target_names=target_names))
print("\n" + "="*40 + " TF-IDF + LR " + "="*40)
print(classification_report(y_test, y_test_lr_pred, target_names=target_names))
print("="*80)

# ==============================================================================
# BAGIAN 7: VISUALISASI CONFUSION MATRIX UNTUK SEMUA MODEL
# ==============================================================================
print("\n" + "="*80)
print("🔍 ANALISIS CONFUSION MATRIX UNTUK SETIAP MODEL")
print("="*80)

class_names = ['Legitimate (Aman)', 'Phishing (Berbahaya)']
models_to_evaluate = [
    {'name': 'BERT (PyTorch)', 'y_true': y_test_bert_true, 'y_pred': y_test_bert_pred},
    {'name': 'LSTM (TensorFlow)', 'y_true': y_test, 'y_pred': y_test_lstm_pred},
    {'name': 'TF-IDF + LR', 'y_true': y_test, 'y_pred': y_test_lr_pred}
]

fig, axes = plt.subplots(1, 3, figsize=(22, 6))
fig.suptitle('Perbandingan Confusion Matrix pada Test Set', fontsize=20, y=1.02)

for i, model_info in enumerate(models_to_evaluate):
    ax = axes[i]
    cm = confusion_matrix(model_info['y_true'], model_info['y_pred'])
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names, ax=ax, cbar=False)
    ax.set_title(f'Model: {model_info["name"]}', fontsize=16)
    ax.set_xlabel('Label Prediksi (Predicted)', fontsize=12)
    ax.set_ylabel('Label Asli (Actual)', fontsize=12)
    tn, fp, fn, tp = cm.ravel()
    print(f"\n--- Confusion Matrix untuk {model_info['name']} ---")
    print(f"  True Negative (TN): {tn}  -> Aman, diprediksi Aman")
    print(f"  False Positive (FP): {fp} -> Aman, diprediksi Phishing")
    print(f"  False Negative (FN): {fn} -> Phishing, diprediksi Aman")
    print(f"  True Positive (TP): {tp}  -> Phishing, diprediksi Phishing")
    print("-" * 40)

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
print("\n✅ Analisis Confusion Matrix selesai.")
print("="*80)
```

```python
# ==============================================================================
# BAGIAN 8: MENYIMPAN MODEL TERBAIK (BERT) UNTUK DEPLOYMENT
# ==============================================================================
print("\n" + "="*80)
print("💾 MENYIMPAN MODEL BERT DAN TOKENIZER")
print("="*80)

# Tentukan direktori untuk menyimpan model
model_save_path = "./best_bert_phishing_classifier"

# Simpan model, tokenizer, dan file konfigurasi
# Ini akan menyimpan semua file yang diperlukan untuk menggunakan model kembali
bert_model_pytorch.save_pretrained(model_save_path)
bert_tokenizer.save_pretrained(model_save_path)

print(f"✅ Model dan tokenizer berhasil disimpan di direktori: {model_save_path}")

# Tampilkan isi dari direktori untuk memverifikasi
print("\n--- Isi Direktori Model ---")
!ls -l {model_save_path}
print("="*80)

# ==============================================================================
# BAGIAN 9: UJI COBA MEMUAT KEMBALI MODEL DAN MELAKUKAN PREDIKSI
# (Simulasi apa yang akan Anda lakukan di VS Code)
# ==============================================================================
print("\n" + "="*80)
print("🚀 MENGUJI PEMUATAN ULANG MODEL DAN FUNGSI PREDIKSI")
print("="*80)

# --- Impor ulang library yang diperlukan untuk simulasi ---
# (Di program asli, ini akan diimpor di bagian paling atas)
from transformers import BertTokenizer, BertForSequenceClassification
import torch

# --- Fungsi untuk memuat model yang sudah disimpan ---
def load_model_and_tokenizer(model_path):
    """Memuat tokenizer dan model BERT dari direktori yang sudah disimpan."""
    print(f"Memuat model dari: {model_path}")
    tokenizer = BertTokenizer.from_pretrained(model_path)
    model = BertForSequenceClassification.from_pretrained(model_path)

    # Pindahkan model ke GPU jika tersedia
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)
    model.eval() # Set model ke mode evaluasi (sangat penting!)

    print(f"✅ Model dan tokenizer berhasil dimuat ke perangkat: {device}")
    return model, tokenizer, device

# --- Fungsi untuk memprediksi satu email ---
def predict_phishing(email_text, model, tokenizer, device):
    """
    Memprediksi apakah sebuah email adalah phishing atau tidak.
    """
    # 1. Bersihkan teks (menggunakan fungsi clean_text yang sudah didefinisikan sebelumnya)
    cleaned_text = clean_text(email_text)

    # 2. Tokenisasi teks
    inputs = tokenizer(
        cleaned_text,
        return_tensors="pt",
        truncation=True,
        padding=True,
        max_length=256
    )

    # 3. Pindahkan input ke perangkat yang sama dengan model (GPU/CPU)
    inputs = {k: v.to(device) for k, v in inputs.items()}

    # 4. Lakukan prediksi
    with torch.no_grad(): # Non-aktifkan perhitungan gradien untuk inferensi yang lebih cepat
        outputs = model(**inputs)

    # 5. Ambil hasil prediksi
    logits = outputs.logits
    probabilities = torch.nn.functional.softmax(logits, dim=1)
    predicted_class_id = torch.argmax(probabilities, dim=1).item()
    confidence_score = probabilities[0, predicted_class_id].item()

    # 6. Ubah ID kelas menjadi label yang mudah dibaca
    label_map = {0: "Legitimate (Aman)", 1: "Phishing (Berbahaya)"}
    prediction_label = label_map[predicted_class_id]

    return {
        "prediction": prediction_label,
        "confidence": f"{confidence_score:.2%}"
    }

# --- Contoh Penggunaan dan Pengujian ---
# Muat model yang baru saja kita simpan
loaded_model, loaded_tokenizer, device = load_model_and_tokenizer(model_save_path)

# Ambil beberapa contoh email dari dataset test untuk diuji
test_samples = final_dataset['test'].select(range(3))

print("\n--- MELAKUKAN PREDIKSI PADA CONTOH EMAIL ---")
for i, sample in enumerate(test_samples):
    sample_text = sample['text']
    true_label = sample['email_type']

    print(f"\n--- Contoh Email ke-{i+1} ---")
    print(f"Label Asli: {true_label}")
    print(f"Teks (100 karakter pertama): {sample_text[:100]}...")

    # Lakukan prediksi menggunakan fungsi kita
    result = predict_phishing(sample_text, loaded_model, loaded_tokenizer, device)

    print("--- Hasil Prediksi Model ---")
    print(f"  Prediksi: {result['prediction']}")
    print(f"  Confidence: {result['confidence']}")
    print("-" * 40)

print("\n✅ Pengujian model yang dimuat ulang selesai.")
print("="*80)
```
